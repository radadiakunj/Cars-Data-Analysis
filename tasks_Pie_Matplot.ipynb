**Task 1 : Plot a pie chart of the "gender" column :**

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('train1.csv')

# Count the number of occurrences of each gender
gender_counts = data['Sex'].value_counts()

# Create a pie chart
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')

# Add a title
plt.title('Gender Distribution')

# Show the plot
plt.show()
```

**Task 2 : Plot a bar graph of the number of alive people using the "survive" column :**

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('train1.csv')

# Count the number of occurrences of each "survive" value
survive_counts = data['Survived'].value_counts()

# Create a bar graph
plt.bar(survive_counts.index, survive_counts)

# Add labels and a title
plt.xlabel('Survival')
plt.ylabel('Count')
plt.title('Number of Survivors')

total = sum(survive_counts)
for i, count in enumerate(survive_counts):
    percentage = count / total * 100
    plt.text(i, count, f'{percentage:.1f}%', ha='center', va='bottom')

# Show the plot
plt.show()
```

**Task 3 : Plot unique values of the "cabin" column and create a subplot with the "cabin" and "gender" columns :**

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('train1.csv')

# Plot unique values of "cabin" column
data['Cabin'].value_counts().plot(kind='bar')
plt.xlabel('Cabin')
plt.ylabel('Count')
plt.title('Unique Cabin Values')

# Create a subplot of "cabin" and "gender" columns
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
data['Cabin'].value_counts().plot(kind='bar')
plt.xlabel('Cabin')
plt.ylabel('Count')
plt.title('Cabin')

plt.subplot(1, 2, 2)
data['Sex'].value_counts().plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender')

# Adjust layout and spacing
plt.tight_layout()

# Show the plot
plt.show()
```

**Task 4 : Plot a scatter graph using the "age" and "Id" columns :**

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('train1.csv')

# Plot a scatter graph of "PassengerId" and "Age" columns
plt.scatter(data['PassengerId'], data['Age'])

# Add labels and a title
plt.xlabel('PassengerId')
plt.ylabel('Age')
plt.title('Scatter Plot : Age vs PassengerId')

# Show the plot
plt.show()
```
